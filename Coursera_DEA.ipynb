```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('coursea_data.csv')
```

```python
df
```

```python
print(df.dtypes)
```

```python
#replacing k with 000
def transform(co):
    data=co.split(",")
    transformed_data=[]
    for i in data:
        if "k" in i:
            i=i.replace("k","")
            i=float(i)*1000
        transformed_data.append(i)
    return",".join([str(elem) for elem in transformed_data])
```

```python
df["students_enrolled"]=df["students_enrolled"].apply(transform)
df
```

```python
#replacing m with 000000
def transform(co):
    data=co.split(",")
    transformed_data=[]
    for i in data:
        if "m" in i:
            i=i.replace("m","")
            i=float(i)*1000000
        transformed_data.append(i)
    return",".join([str(elem) for elem in transformed_data])
```

```python
df["students_enrolled"]=df["students_enrolled"].apply(transform)
df
```

```python
#typecasting object to integer
df["students_enrolled"]=pd.to_numeric(df["students_enrolled"], downcast='integer')
```

```python
#rounding the Rating column
df.round({"rating":1})
```

```python
#approximating the rating data
df.loc[(df['rating'] >= 1) & (df['rating'] <= 1.3), 'revised_rating'] = 1
df.loc[(df['rating'] >= 1.4) & (df['rating'] <= 1.7), 'revised_rating'] = 1.5
df.loc[(df['rating'] >= 1.8) & (df['rating'] <= 2.3), 'revised_rating'] = 2
df.loc[(df['rating'] >= 2.4) & (df['rating'] <= 2.7), 'revised_rating'] = 2.5
df.loc[(df['rating'] >= 2.8) & (df['rating'] <= 3.3), 'revised_rating'] = 3
df.loc[(df['rating'] >= 3.4) & (df['rating'] <= 3.7), 'revised_rating'] = 3.5
df.loc[(df['rating'] >= 3.8) & (df['rating'] <= 4.3), 'revised_rating'] = 4
df.loc[(df['rating'] >= 4.4) & (df['rating'] <= 4.7), 'revised_rating'] = 4.5
df.loc[(df['rating'] >= 4.8) & (df['rating'] <= 5), 'revised_rating'] = 5
```
